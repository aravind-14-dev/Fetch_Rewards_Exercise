# Exploring Receipts data

```python
df = pd.read_json('receipts.json',lines=True)
df.head()
```

# Converting the dictionary data to a single value in '_id' column

```python
df['_id'] = df['_id'].apply(pd.Series)
df.head()
```

# Checking if all the records are unique

```python
df['_id'].is_unique # All records are unique
```

# Verifying the datatypes

```python
df.info()
```

There are 15 columns and 1119 records.

Data Quality Issue 1:

There are a lot of missing values in "bonuspointsEarned" and "pointsEarned" which cannot be imputated in ths case. But with the business logic behind the points calculation, they can be imputated using regression.

# Checking the statistics of the data

```python
df.describe()
```

# Verifying if the receipts with over 600 items are valid records

```python
df[df['purchasedItemCount'] > 600]
```

Verified with the given JSON that the receipts with over 600 items are valid.

# Verifying if a receipt with 10199 pointsEarned is valid

```python
df[df['pointsEarned'] == 10199.8]
```

Possible Data Quality Issue:

The receipt has 5 bonus points,only 2 items and total spent is 11. I did a research to find out that 1000 points equal to one dollar. So, earning 10000+ points on a total spent of 11 dollars doesn't look right. it might be an invalid record/outlier.
